```python
import pandas as pd

daily = pd.read_csv("../data/processed/daily_sales.csv", parse_dates=["order_date"])

print("Shape:", daily.shape)
print("Date range:", daily["order_date"].min(), "to", daily["order_date"].max())
print("Columns:", daily.columns.tolist())

# Check for gaps in the date index — critical for forecasting later
expected_days = pd.date_range(daily["order_date"].min(), daily["order_date"].max(), freq="D")
missing_days = expected_days.difference(daily["order_date"])
print(f"\nMissing calendar days: {len(missing_days)}")

# Check how many days have zero sales (expected some, but not the majority)
zero_sales_days = (daily["total_sales"] == 0).sum()
print(f"Days with zero sales: {zero_sales_days} out of {len(daily)} ({zero_sales_days/len(daily):.1%})")
```
